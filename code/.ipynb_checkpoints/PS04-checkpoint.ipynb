## Question 2

Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
def dftSum(ys):
    N = ys.shape[0]
    n = numpy.arange(N)
    k = n.reshape((N,1))
    M = numpy.exp(-2j*numpy.pi*k*n/N)
    X = numpy.dot(M, ys)
    return X
```

```python
sig = thinkdsp.SinSignal()
wave = sig.make_wave()
spec1 = wave.make_spectrum()
spec2 = thinkdsp.Spectrum(hs=numpy.fft.fft(wave.ys),framerate=wave.framerate)
spec3 = thinkdsp.Spectrum(hs=dftSum(wave.ys),framerate=wave.framerate)
print(numpy.allclose(spec2.hs,spec3.hs))
%timeit dftSum(wave.ys)
%timeit numpy.fft.fft(wave.ys)
```

## Question 3

Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def cooleytukeyFFT(wavearray):
    def wrapper(ys):
        N = ys.shape[0]
        
        if N <= 16:
            return dftSum(ys)
        evenpart = cooleytukeyFFT(ys[::2])
        oddpart = cooleytukeyFFT(ys[1::2])
        
        combined = [0] * N
        for m in xrange(N//2):
            combined[m] = evenpart[m] + numpy.exp(-2j*numpy.pi*m/N) * oddpart[m]
            combined[m + N//2] = evenpart[m] - numpy.exp(-2j*numpy.pi*m/N) * oddpart[m]
        return numpy.array(combined)
    
    length = wavearray.shape[0]
    w = zero_pad(wavearray,int(2**(numpy.ceil(numpy.log2(length)))))
    return wrapper(w)
```

```python
sig = thinkdsp.SinSignal()
wave3 = sig.make_wave()
spec1 = wave3.make_spectrum()
spec2 = thinkdsp.Spectrum(hs=numpy.fft.fft(zero_pad(wave3.ys,2**(numpy.ceil(numpy.log2(len(wave3.ys)))))),framerate=wave3.framerate)
spec3 = thinkdsp.Spectrum(hs=cooleytukeyFFT(wave3.ys),framerate=wave3.framerate)
print(numpy.allclose(spec2.hs,spec3.hs))
%timeit cooleytukeyFFT(wave3.ys)
%timeit numpy.fft.fft(wave3.ys)
```

## Question 4

Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
'''Alan's autocorr from Chapter 5'''

def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)

ns = 2 ** numpy.arange(6, 14)
print(ns)
```

```python
'''Timed result'''
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o autocorr(thinkdsp.Wave(ys,framerate = noise.framerate))
    results.append(result)
    
bests = [result.best for result in results]

x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

```python
'''Numpy autocorrelate'''
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o numpy.correlate(ys,ys)
    results.append(result)
    
bests = [result.best for result in results]

x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

```python
''' Chapter 8 implementation'''
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
'''Timed Result'''
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o fft_autocorr(ys)
    results.append(result)
    
bests = [result.best for result in results]

x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
print('estimated slope', t[0])

thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log', legend=False)
```

From the results, I'm not sure if numpy's autocorrelate uses FFT but it is likely because of the ease of doing autocorrelation from the FFT.
