```python
from __future__ import print_function, division

import numpy
import thinkdsp
import thinkplot

%matplotlib inline


sin = []
for i in range(100):
    if i%2 == 0:
        sin.append(thinkdsp.SinSignal(freq = 220+220*(i), amp = 1/(i+1)))
mix = sum(sin)
wave = mix.make_wave()
wave.normalize()
wave.apodize()
mix.plot()
wave.make_audio()
```

To change the waveform shape, modify the 'if' statement to filter out odd or even harmonics and change the amplitude dropoff ($1/f$, $1/f^2$, $1/f^3$, etc.) Change the signal between Sin and Cos to generate different wave.

```python
wave.make_spectrum().plot()
```
