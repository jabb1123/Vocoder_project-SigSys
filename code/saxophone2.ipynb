This notebook contains an example that explores the missing fundamental phenomenon.

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import random

%matplotlib inline
```

Let's investigate the missing fundamental phenomenon.  I'll load a recording of a saxophone.

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
wave.normalize()
wave.make_audio()
```

The spectrogram shows the harmonic structure over time.

```python
gram = wave.make_spectrogram(seg_length=1024)
gram.plot(high=80)
thinkplot.config(xlabel='time (s)', ylabel='frequency (Hz)', legend=False)
```

To see the harmonics more clearly, I'll take a small segment at the 3 second mark.

```python
duration = 0.03
segment = wave.segment(start=3.0, duration=duration)
segment.plot()
```

And let's see the spectrum:

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude')
```

The fundamental frequency is near 200 Hz, but the dominant component is around 600 Hz.  The perceived pitch is the fundamental frequency.

To investigate the autocorrelation of this signal, we can plot shifted segments together:

```python
def plot_shifted(wave, shift=0.002, start=3.0):
    duration = 0.015

    thinkplot.preplot(2)
    segment1 = wave.segment(start=start, duration=duration)
    segment1.plot(linewidth=2, alpha=0.8)

    segment2 = wave.segment(start=start-shift, duration=duration)
    segment2.plot(linewidth=2, alpha=0.4)

    corr = segment1.corr(segment2)
    text = r'$\rho =$ %.2g' % corr
    thinkplot.text(0.0005, -0.8, text)
    thinkplot.config(xlabel='time (s)', ylim=[-1, 1])

plot_shifted(wave, 0.0002)
```

With a small shift the segments are still moderately correlated.  As the shift increases, the correlation falls for a while, then rises again, peaking when the shift equals the period of the signal.

You can use the following interaction to search for the shift that maximizes correlation:

```python
from IPython.html.widgets import interact, fixed
from IPython.html import widgets

end = 0.006
slider1 = widgets.FloatSliderWidget(min=0, max=end, step=end/40, value=0)
slider2 = widgets.FloatSliderWidget(min=0.1, max=0.5, step=0.05, value=3.0)
interact(plot_shifted, wave=fixed(wave), shift=slider1, start=slider2)
None
```

We can use `numpy.correlate` to compute the autocorrelation function.  I'll use a longer segment so we can hear it.

```python
segment = wave.segment(start=3.0, duration=0.5)
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=1000)
```

The following function computes the autocorrelation function, selects the second half (which corresponds to positive lags) and normalizes the results:

```python
def autocorr(segment):
    corrs = numpy.correlate(segment.ys, segment.ys, mode='same')
    N = len(corrs)
    lengths = range(N, N//2, -1)

    half = corrs[N//2:].copy()
    half /= lengths
    half /= half[0]
    return half
```

```python
corrs = autocorr(segment)
thinkplot.plot(corrs[:500])
thinkplot.config(xlabel='lag', ylabel='correlation', ylim=[-1.05, 1.05])
```

The highest peak (other than 0) is between lag=200 and 250.

We can use `argmax` to find the index of that peak:

```python
low, high = 200, 250
lag = numpy.array(corrs[low:high]).argmax() + low
lag
```

We can convert from an index to a time in seconds:

```python
period = lag / segment.framerate
period
```

Given the period in seconds, we can compute frequency:

```python
frequency = 1 / period
frequency
```

So the peak in the autocorrelation function corresponds to the perceived pitch, which is close to G#3/Ab3.

For comparison, here's a xxx at 208 Hz.

```python
thinkdsp.SawtoothSignal(freq=208).make_wave(duration=0.5).make_audio()
```

Now let's see what happens when we start filtering our harmonics.  Here's what it looks like with a high-pass filter:

```python
spectrum2 = segment.make_spectrum()
spectrum2.high_pass(450)
spectrum2.plot(high=1000)
```

And here's what it sounds like.

```python
segment2 = spectrum2.make_wave()
segment2.make_audio()
```

The perceived pitch is still 208 Hz, even though there is no power at 208 or 416 Hz.  This is the missing fundamental phenomenon.

The autocorrelation function shows why we still hear 208 Hz.

```python
corrs = autocorr(segment2)
thinkplot.plot(corrs[:500])
thinkplot.config(xlabel='lag', ylabel='correlation', ylim=[-1.05, 1.05])
```

Because of the higher harmonics, the signal has a stronger autocorrelation at 208 Hz (which corresponds to the third hump) than at 416 or 624 Hz (the previous two humps).

If we filter out some of the high harmonics, the effect is diminished.

```python
spectrum3 = segment.make_spectrum()
spectrum3.high_pass(450)
spectrum3.low_pass(1000)
spectrum3.plot(high=1000)
```

Now the perceived pitch is ambiguous, at least to my hear.

```python
segment3 = spectrum3.make_wave()
segment3.make_audio()
```

If we look at the autocorrelation function, the third hump is still higher than the previous two, but not by much.

```python
corrs = autocorr(segment3)
thinkplot.plot(corrs[:500])
thinkplot.config(xlabel='lag', ylabel='correlation', ylim=[-1.05, 1.05])
```

Finally, if we remove all of the harmonics, we are left with something pretty close to a pure sinusoid, and the perceived pitch pops up to 624 Hz.

```python
spectrum4 = segment.make_spectrum()
spectrum4.high_pass(450)
spectrum4.low_pass(750)
spectrum4.plot(high=1000)
```

```python
segment4 = spectrum4.make_wave()
segment4.make_audio()
```

For comparison, here is a pure sinusoid at 624 Hz.

```python
thinkdsp.SinSignal(freq=624).make_wave(duration=0.5).make_audio()
```

And the autocorrelation function now shows no visibile difference between the first three humps.

```python
corrs = autocorr(segment4)
thinkplot.plot(corrs[:500])
thinkplot.config(xlabel='lag', ylabel='correlation', ylim=[-1.05, 1.05])
```

In conclusion, it looks like the perceived pitch is consistent with the results of autocorrelation, and contrary to what we would expect based on spectral decomposition.  Although it seems clear that our hearing uses some kind of spectral decomposition, the missing fundamental phenomenon suggests that somewhere in the auditory processing pathway, something like autocorrelation is also going on.

As an aside, the autocorrelation structure depends on the phase of the harmonics.  If we shuffle the phases, we get something that sounds different from the original, and the pitch is ambiguous (at least to my ear).

```python
spectrum5 = spectrum.copy()
magnitude = numpy.abs(spectrum5.hs)
angle = numpy.angle(spectrum5.hs)
random.shuffle(angle)
i = complex(0, 1)
spectrum5.hs = magnitude * numpy.exp(i * angle)
segment5 = spectrum5.make_wave()
segment5.apodize()
segment5.make_audio()
```

```python
corrs = autocorr(segment5)
thinkplot.plot(corrs[:500])
thinkplot.config(xlabel='lag', ylabel='correlation', ylim=[-1.05, 1.05])
```
