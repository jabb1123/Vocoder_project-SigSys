6)The Sawtooth implementation.

```python
class SawtoothSignal (thinkdsp.Sinusoid):
    
    def evaluate (self,time_array):
        cycle = self.freq * time_array +self.offset / (numpy.pi*2)
        frac, junk=numpy.modf(cycle)
        return frac
        
```

```python
saw1 = SawtoothSignal(freq = 440)
saw2 = SawtoothSignal(freq = 880)
mix= saw1 + saw2
wave = mix.make_wave(duration=1, start=0, framerate=11025)
wave.normalize()
wave.apodize()
mix.plot()
wave.make_audio()
```

```python
wave.make_spectrum().plot()
```

10)The sin signal would contain even and odd harmonics while having the $1/f^2$ drop off.  The signal would have to be continuous while the derivative of the signal is discontinuous.  It would aslo have non-symetric wave forms.

```python
sin = []
for i in range(10):
    sin.append(thinkdsp.SinSignal(freq = 440+440*i, amp = 1/(i+1)**2))
mix = sum(sin)
wave = mix.make_wave()
wave.normalize()
wave.apodize()
mix.plot()
wave.make_audio()
```

```python
wave.make_spectrum().plot()
```
