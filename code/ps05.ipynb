So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
window1 = numpy.fft.fft(scipy.signal.gaussian(M=11025, std=2))
sig = thinkdsp.SquareSignal()
wave = sig.make_wave()
angle = numpy.fft.fft(wave.ys)
inverse_signal = numpy.fft.ifft(angle * window1)
phase_offset = numpy.angle(inverse_signal)
thinkplot.plot(angle,phase_offset)
```

```python
window2 = numpy.fft.fft(scipy.signal.boxcar(M=11025))
sig = thinkdsp.SquareSignal()
wave = sig.make_wave()
angle = numpy.fft.fft(wave.ys)
inverse_signal = numpy.fft.ifft(angle * window2)
phase_offset = numpy.angle(inverse_signal)
thinkplot.plot(angle,phase_offset)
```

## Question 4

Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

signal = thinkdsp.SquareSignal()
wave = signal.make_wave()
ys = numpy.convolve(wave.ys, zero_pad([1,-1], len(wave.ys)), mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

```python
dft_window = numpy.fft.rfft(zero_pad([1,-1],len(wave.ys)))
thinkplot.plot(abs(dft_window))
```

```python
signal = thinkdsp.SquareSignal()
wave = signal.make_wave()
ys = numpy.convolve(wave.ys, zero_pad([-0.5, 2, -1.5], len(wave.ys)), mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

```python
dft_window = numpy.fft.rfft(zero_pad([-.5,2,-1.5],len(wave.ys)))
thinkplot.plot(abs(dft_window))
```

```python
signal = thinkdsp.SquareSignal()
wave = signal.make_wave()
ys = numpy.convolve(wave.ys, zero_pad([1,-2, 1], len(wave.ys)), mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
spectrum2.plot()
```

```python
dft_window = numpy.fft.rfft(zero_pad([1,-2,1], len(wave.ys)))
thinkplot.plot(abs(dft_window))
```
